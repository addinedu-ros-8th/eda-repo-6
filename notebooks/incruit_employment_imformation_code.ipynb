```python
##시작

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
from tqdm import tqdm
from selenium.webdriver import ActionChains
from selenium.common.exceptions import StaleElementReferenceException, TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# options = webdriver.ChromeOptions()

# url = "https://www.incruit.com/?utm_source=google&utm_medium=cpc&utm_campaign=paid_keyword_google_pc&utm_term=%EC%9D%B8%EC%BF%A0%EB%A5%B4%ED%8A%B8&epe_vid=103&epe_tcd=40174951&pco=107&gad_source=1&gclid=Cj0KCQiAsaS7BhDPARIsAAX5cSDYfCe6pRxF39g6HDavotc5Q_pqlw0ypbmxvTWsAej0h7_mtEWrYMkaAiX_EALw_wcB"

# #리눅스
# # driver = webdriver.Chrome(service=Service("../driver/chromedriver"))
# #windows
# driver = webdriver.Chrome(service=Service("C:\\Users\\NDY\\Desktop\\dev_ws\\driver\\chromedriver.exe"))

# driver.get(url)

# #리눅스
# # driver = webdriver.Chrome(service=Service("../driver/chromedriver"))
# #windows
driver = webdriver.Chrome(service=Service("C:\\Users\\NDY\\Desktop\\dev_ws\\driver\\chromedriver.exe"))

base_url = "https://search.incruit.com/list/search.asp?col=job&kw=%b7%ce%ba%bf+sw+%b0%b3%b9%df&startno={}&psize=30"

# 딕셔너리 초기화
Employment_Information_data = {
                                "기업명": "",
                                "필수조건" : "",
                                "우대사항" : "",
                                "경력" : "",
                                "학력" : ""
                                
}


# 각 리스트 초기화
company_name_list = []
career_list = []
education_background_list = []

for page_number in range(0, 1560, 30):
    url = base_url.format(page_number)
    driver.get(url)
    time.sleep(1)

    company_info = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CLASS_NAME, "cBbslist_contenst")))
    company_ctn = company_info.find_elements(By.CLASS_NAME, "c_row")

    if company_info:
        for each in tqdm(company_ctn):

            # 로봇 sw 개발 구직회사 찾기
            Employment_Information_Box = each.find_element(By.CLASS_NAME, "cell_mid")
            find_robot = Employment_Information_Box.find_element(By.TAG_NAME, "a")
            
            if "로봇"  in find_robot.text:

                # 기업명 
                company_names = each.find_element(By.CLASS_NAME, "cell_first")
                company_name = company_names.find_element(By.TAG_NAME, "a")

                # 채용 정보 찾기
                Employment_Information_btm = each.find_element(By.CLASS_NAME, "cl_md")
                Employment_Information_span = Employment_Information_btm.find_elements(By.TAG_NAME, "span")

                Employment_Information_span_temp = []
                
                # 채용 정보 리스트에 저장
                for each_Employment_Information_span in Employment_Information_span:
                    Employment_Information_span_temp.append(each_Employment_Information_span.get_attribute("innerHTML"))

                #기업명 리스트에 저장
                company_name_list.append(company_name.get_attribute("innerHTML"))       

                # # 채용 분야
                # job_category_list.append(find_robot.text)
                # print(find_robot.text)

                # 채용정보 리스트에 저장
                career_list.append(Employment_Information_span_temp[0])
                education_background_list.append(Employment_Information_span_temp[1])
                # employment_type_list.append(Employment_Information_span_temp[3])
                

            else:
                continue
           

    else:
        break
Employment_Information_data["기업명"] = company_name_list
Employment_Information_data["경력"] = career_list
Employment_Information_data["학력"] = education_background_list

```
