```python
import pandas as pd
import numpy as np
```

```python
company = pd.read_csv("../../../Downloads/merged_company_info_ver5.csv")
## file link: https://tlacogns-1734568117216.atlassian.net/wiki/x/oxMB
company.info()
```

```python
for idx, row in enumerate(company["설립일"]):
    if row == None:
        continue
    company["설립일"][idx] = str(row)[:4]
company
```

```python
company["사원수"] = company["사원수"].fillna(0)
company["최저연봉"] = company["최저연봉"].fillna(0)
company["최고연봉"] = company["최고연봉"].fillna(0)
company["평균연봉"] = company["평균연봉"].fillna(0)

company = company.astype({"사원수":np.int32,
                        "기업형태":np.int32,
                        "최저연봉":np.int32,
                        "평균연봉":np.int32,
                        "최고연봉":np.int32,
                        "주소":np.int32,
                        "구":np.int32})

company
```

```python
company = company.astype({"사원수":object,
                        "기업형태":object,
                        "최저연봉":object,
                        "평균연봉":object,
                        "최고연봉":object,
                        "주소":object,
                        "구":object})
```

```python
for each in company["매출액"]:
    print(each)
```

```python
for idx, each in enumerate(company["매출액"]):
    if each == "정보없음" or each == None:
        company["매출액"][idx] = None
        continue
    if each.find("억")>0:
        company["매출액"][idx] = str(int(float(each[:-1])*10000))
    elif each.find("만")>0:
        company["매출액"][idx] = each.replace(',', '')[:-1]
company    
```

```python
for each in company["매출액"]:
    print(each)
```

```python
for idx, each in company.iterrows():
    if each["구"] == 0:
        company["구"][idx] = 96
    if each["주소"] == 0:
        company["주소"][idx] = 17
    if each["기업형태"] == 0:
        company["기업형태"][idx] = 11

company
```
